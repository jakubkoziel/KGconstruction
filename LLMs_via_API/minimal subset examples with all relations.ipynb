```python
def greedy_set_cover_indices(list_of_lists):
    # Convert lists to sets
    sets = [set(lst) for lst in list_of_lists]
    universe = set().union(*sets)
    remaining = set(universe)
    selected = []
    available_indices = list(range(len(sets)))  # Track remaining indices

    while remaining:
        # Find the index covering most remaining elements
        best_idx = -1
        best_cover = 0
        for idx in available_indices:
            coverage = len(sets[idx] & remaining)
            if coverage > best_cover:
                best_cover = coverage
                best_idx = idx

        if best_idx == -1:  # No coverage possible
            break

        selected.append(best_idx)
        remaining -= sets[best_idx]
        available_indices.remove(best_idx)

    return selected

```

```python
def weighted_greedy_set_cover(list_of_lists, counts):
    # Convert lists to sets and prepare tracking
    sets = [set(lst) for lst in list_of_lists]
    universe = set(counts.keys())
    remaining = set(universe)
    selected_indices = []
    available_indices = list(range(len(sets)))

    coverage_data = []
    total_weight = sum(counts.values())

    while remaining and available_indices:
        # Find the set with maximum weighted coverage
        best_idx = -1
        best_weight = 0
        best_coverage = set()

        for idx in available_indices:
            current_coverage = sets[idx] & remaining
            current_weight = sum(counts[e] for e in current_coverage)
            if current_weight > best_weight:
                best_weight = current_weight
                best_idx = idx
                best_coverage = current_coverage

        if best_idx == -1:
            break  # No more coverage possible

        # Update tracking
        selected_indices.append(best_idx)
        remaining -= best_coverage
        available_indices.remove(best_idx)

        # Track coverage progress
        coverage_data.append({
            'num_sets': len(selected_indices),
            'covered_weight': total_weight - sum(counts[e] for e in remaining),
            'remaining_weight': sum(counts[e] for e in remaining),
            'excluded_elements': list(remaining)
        })

    return selected_indices, coverage_data
```

```python
selected_indices, coverage_history = weighted_greedy_set_cover(rel_occurrences, counts)

# Print results
print("Selected set indices:", selected_indices)
print("\nCoverage progression:")
for step in coverage_history:
    print(f"With {step['num_sets']} sets: "
          f"Covered {step['covered_weight']:,} ({step['covered_weight']/sum(counts.values()):.1%}), "
          f"Remaining weight: {step['remaining_weight']:,}")
```

```python
def revert_relation_mapping(response):
    with open('../data/DocRED/rel_info.json', 'r', encoding='utf-8') as file:
        relations = json.load(file)
    relations = {v: k for k, v in relations.items()}

    response_remapped = []
    for rel in response:
        if rel["r"] in relations:
            new_rel = rel.copy()
            new_rel["r"] = relations[rel["r"]]
            response_remapped.append(new_rel)
        else:
            print(f"WARNING: Relation {rel['r']} not found in relation_dict")
    return response_remapped
```

```python
import os
v = 'v7'
base = f're_entities_separately/docred/dev/{v}_not_renamed'
for model in ['deepseek-chat', 'deepseek-reasoner', 'gpt-4o-mini']:
    files = os.listdir(os.path.join(base, model))
    # print(files)
    for f in files:
        path = os.path.join(base, model, f)
        #print(path)
        response = json.load(open(path, encoding='utf-8'))
        response = revert_relation_mapping(response=response)
        #print(path.replace(f'{v}_not_renamed', v))
        with open(path.replace(f'{v}_not_renamed', v), 'w', encoding='utf-8') as f:
            json.dump(response, f, indent=4, ensure_ascii=False)
        #break
```
